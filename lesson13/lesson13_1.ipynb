```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y, 'bo-.')

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
ax_1 = figure.add_subplot(1, 3, 1)
ax_2 = figure.add_subplot(1, 3, 2) 
ax_3 = figure.add_subplot(1, 3, 3)

ax_1.set_title("1")
ax_2.set_title("2")
ax_3.set_title("3")

plt.show()

```

```python
import matplotlib.pyplot as plt
student1:dict[str,int] = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2:dict[str,int] = {'chinese':90, 'english':60, 'math':85, 'history':62, 'discover':73}
subject = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subject,score1)
axes.plot(subject,score2)
plt.show()
```

```python

#window字型路徑
#font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
#=======
#mac字型路徑
#font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#=======
#linux
#安裝字型
#sudo apt-get update -y
#sudo apt-get install -y fonts-wqy-zenhei
#linux字型路徑
#font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
student1:dict[str,int] = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2:dict[str,int] = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
subject = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subject,score1)
axes.plot(subject,score2)
axes.set_xticklabels(subject,fontproperties=font)
plt.show()
```
